Primero saco la cuenta de cantidad total de varones, mujeres y personas sin especificar fallecidos.

```
df_cantdef = dfagg.groupby(['SEXO'])['CUENTA'].sum().to_frame().join(dfsex.set_index('CODIGO'), on='SEXO').reset_index()
df_cantdef.rename(columns = {'VALOR':'Sexo'}, inplace = True)
df_cantdef = df_cantdef[['CUENTA','Sexo']]
df_cantdef
```

Ahora agrupo las muertes por sexo y causa.

```
df_bycause = dfagg.groupby(['SEXO','CAUSA'])['CUENTA'].sum().to_frame().join(dfcod.set_index('CODIGO'), on='CAUSA')
df_bycause = df_bycause.reset_index()

new_columns3 = df_bycause.columns.values
new_columns3[3] = "Causa de muerte"
df_bycause.columns = new_columns3

df_bycause2 = df_bycause.join(dfsex.set_index('CODIGO'), on='SEXO')

new_columns4 = df_bycause2.columns.values
new_columns4[4] = "Sexo"
df_bycause2.columns = new_columns4

df_bycause3 = df_bycause2[['Sexo','Causa de muerte','CUENTA']]
df_bycause3
```

Ahora tomo las 10 causas más comunes de muerte en varones, sacando el porcentaje del total de varones que fallecieron debido a la misma.

```
#Varones
df_bycause4 = df_bycause3.loc[df_bycause3['Sexo'] == 'Varón']
df_bycause4 = df_bycause4.sort_values(by='CUENTA', ascending=False)
df_bycause4 = df_bycause4.head(10)
df_bycause4['PORCENTAJE'] = df_bycause4['CUENTA']/1535170*100
df_bycause4
```

```
#Varones
alt.Chart(df_bycause4).mark_bar(color='#4c78a8').encode(
    x=alt.X('PORCENTAJE:Q', 
          title='Porcentaje'),
    y=alt.Y('Causa de muerte:N', 
          title='Causa de muerte', 
          axis=alt.Axis(labelAngle=0),
          sort='-x')
)
```

Vemos que la causa de defunción más común en varones es **insuficiencia cardíaca** (7.36%), seguida de **neumonía organismo no especificado** (6.70%) e **infarto agudo del miocardio** (5.59%).

```
#Mujeres
df_bycause5 = df_bycause3.loc[df_bycause3['Sexo'] == 'Mujer']
df_bycause5 = df_bycause5.sort_values(by='CUENTA', ascending=False)
df_bycause5 = df_bycause5.head(10)
df_bycause5['PORCENTAJE'] = df_bycause5['CUENTA']/1435227*100

df_bycause5
```

```
#Mujeres
alt.Chart(df_bycause5).mark_bar(color='#e45756').encode(
    x=alt.X('PORCENTAJE:Q', 
          title='Porcentaje'),
    y=alt.Y('Causa de muerte:N', 
          title='Causa de muerte', 
          axis=alt.Axis(labelAngle=0),
          sort='-x')
)
```

Vemos que la causa de defunción más común en mujeres es **insuficiencia cardíaca** (9.77%), seguida de **neumonía organismo no especificado** (8.12%) y **otras causas mal definidas y las no especificadas** (4.82%). Las dos primeras son las mismas que en el caso de los varones, aunque ambas son más frecuentes en mujeres. La tercera es distinta, aunque es la 4ta más frecuente en varones, al igual que la 3ra más frecuente en varones es la 4ta más frecuente en mujeres (están invertidas).

Cabe destacar que algunas causas de muerte difieren bastante según el sexo, como la 5ta causa de muerte más común en hombres, que es **tumor maligno de los bronquios y del pulmón**, bastante menos frecuente en mujeres (3.75% versus 1.85%), o la 6ta causa más frecuente en mujeres, inexistente en varones, que es **tumor maligno de la mama** (3.58%).

Por último veamos las causas de muerte más frecuentes en individuos cuyo sexo no está especificado, con el fin de analizar si el sexo del cuerpo era indistinguible al momento de determinar la causa de muerte (por ejemplo por quemaduras, desmembramiento, o similar), o si simplemente no se registró el sexo del individuo.

```
#Sin especificar
df_bycause6 = df_bycause3.loc[df_bycause3['Sexo'] == 'Sin especificar']
df_bycause6 = df_bycause6.sort_values(by='CUENTA', ascending=False)
df_bycause6 = df_bycause6.head(10)
df_bycause6['PORCENTAJE'] = df_bycause6['CUENTA']/3449*100

df_bycause6
```

```
#Sin especificar
alt.Chart(df_bycause6).mark_bar(color='#f58518').encode(
    x=alt.X('PORCENTAJE:Q', 
          title='Porcentaje'),
    y=alt.Y('Causa de muerte:N', 
          title='Causa de muerte', 
          axis=alt.Axis(labelAngle=0),
          sort='-x')
)
```

Vemos que las principales causas de muerte son similares a las de varones y mujeres, lo cual sugiere que simplemente el registro del sexo fue deficiente al momento de la muerte.
